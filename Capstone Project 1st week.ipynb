# Business Problem

Businessman wants to open a distribution center from where he can supply products easily. This products usually are Rice, Beans and spices. He is looking for a place where he can buy a storage facility. The place should be in a place where there are many restaurants nearby. For this he will compare few cities to decide where his business can be profitable for him. 

In this project , we will help him explore the cities where he can find more customers and can reach them easily. If requested then more information about restaurant menu will be provided so he can even more easily make descions as fast food restaurant will not be his main customers. So first thing is to provide him number of restaurents in each city then category of restaurans. So he can make decision more precisly. 

# Data

In this case we will consider two of the largest cities of Sweden. Mälmo and Gothenburg. We will first compare the number of restaurants in both cities. Then we will try to categorise fast food and other categories in order to make clear judgment.
